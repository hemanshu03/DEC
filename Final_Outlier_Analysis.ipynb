# Function to calculate z-scores of an array of Data

```python
def z_score(x):
	mean=np.mean(x)
	std=np.std(x)

	scaled_data=[]
	for i in x:
		z = (i - mean)/std
		scaled_data.append(z)
	return(scaled_data)
```

# Outliers
Outliers are data points that are significantly different from the majority of data points in a dataset. They are observations that deviate markedly from the rest of the data and can have a substantial impact on statistical analyses and machine learning models.

# Reasons for Outliers to exist in the data:
* Data Collection Errors: Outliers may emerge due to errors in data collection, including instrument inaccuracies and data entry mistakes.

* **Natural Variability:** Some data points represent genuine extremes caused by infrequent or exceptional events.

* **Data Input Mistakes:** Outliers can result from typographical or data input errors.

* **Sampling Variability:** Inadequate sample representation or small sample sizes can lead to the inclusion of extreme values by chance.

* **Heterogeneous Data:** Outliers may arise when there are multiple subgroups in the data with distinct characteristics.

* **Data Transformation Effects:** Poorly applied data transformations or highly skewed distributions can introduce outliers.

* **Experimental Fluctuations:** In scientific experiments, unexpected variations or errors can produce outliers.

* **Rare Occurrences:** Outliers might indicate rare but significant events or conditions within the dataset.

* **Data Manipulation:** Intentional or accidental introduction of outliers, perhaps for testing or synthetic data generation.

* **Missing Data Impact:** When data is missing for one variable, extreme values in other variables can appear as outliers.

* **Temporal Changes:** Changes in underlying processes in time-series data can generate outliers.

# Finding the Outliers:
**1. Visual Inspection**

**Box Plots:**
* Box plots provide a visual summary of the distribution of your data.

* Outliers are typically shown as individual data points beyond the "whiskers" of the plot.

* Points outside the whiskers can be considered potential outliers.

**Scatter Plots:**

* Scatter plots are useful for visualizing relationships between two continuous variables.

* Outliers often appear as data points that are far away from the main cluster of points.

**2. Descriptive Statistics:**

**Z_Score:**
The Z-score, also known as the standard score or standardized score, is a statistical measure that quantifies how many standard deviations a data point is from the mean (average) of a dataset. It helps in assessing the relative position of a data point within a distribution.

The formula for calculating the Z-score of an individual data point, x, in a dataset with mean (μ) and standard deviation (σ) is:
$$ Z = \frac{x - \mu}{\sigma} $$

where:

* Z: The Z-score, which represents how many standard deviations a data point is from the mean.

* x: The individual data point you want to calculate the Z-score for.

* μ (Mu): The mean (average) of the dataset.

* σ (Sigma): The standard deviation, which measures the amount of variation or dispersion in the dataset.

**IQR[Inter Quartile Range]:**

The Interquartile Range (IQR) is a measure of statistical dispersion that quantifies the spread or variability of a dataset. It is defined as the range between the first quartile (Q1) and the third quartile (Q3) of the data when the data is sorted in ascending order. The IQR is used to describe the middle 50% of the data and is less sensitive to extreme values or outliers compared to the range or standard deviation.

**2. Visualisation Tools:**

* **Heatmap:** A heatmap is a graphical representation of data in a two-dimensional matrix, where values in the matrix are represented using colors. Heatmaps are commonly used to visualize data that has two dimensions, such as a table of values or a matrix, where each cell's color represents the magnitude or intensity of the data at that specific location.

* **Box Plots:** A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's distribution. It provides a visual summary of the central tendency, spread, and potential outliers in the data. Box plots are particularly useful for comparing distributions between multiple groups or datasets.

* **Histograms:** A histogram is a graphical representation of the distribution of a dataset. It provides a visual summary of the frequency or count of values within specific intervals, called bins, along a single axis. Histograms are useful for understanding the shape, central tendency, and spread of data.

# Finding the Outliers in the data using Z-Score

```python
def outliers(x):
		outliers=[]
		mean=np.mean(x)
		std=np.std(x)
		for i in x:
			z = (i - mean)/std
			if (np.abs(z) > 3):
				outliers.append(i)
		return(outliers)
```

```python
a = outliers(data)
a
```

# Finding the Outliers in the data usiing Inter Quartile Range [IQR]

```python
q1,q3=np.percentile(dataset,[25,75])
```

```python
print('The First Quartile is:',q1)
```

```python
print('The Third Quartile is:',q3)
```

# Calculating IQR

```python
iqr = q3 - q1
print('The Inter Quartile Range is:',iqr)
```

# Calculating the Lower Bound & the Upper Bound

```python
lower_bound = q1 - iqr*1.5
print('Lower Bound of the Data Range is:',lower_bound)
```

```python
upper_bound = q3 + iqr*1.5
print('Upper Bound of the Data Range is:',upper_bound)
```

# Finding Outliers using the Lower & the Upper Bound

```python
Outliers=[]
for i in data:
    if i<lower_bound or i>upper_bound:
        Outliers.append(i)

Outliers
```
