2. Calculate the mean of the counted_pct column

```python
counted_pct_mean = df.counted_pct.mean()
```

```python
print("Mean of COUNT column: ",counted_pct_mean)
```

Impute the missing data in the age column with the mean age value

```python
df.counted_pct.fillna(counted_pct_mean,inplace=True)
```
