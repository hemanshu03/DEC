# Feature normalization
Feature normalization, also known as feature scaling, is a data preprocessing technique used to standardize or rescale the values of features in a dataset. Normalization ensures that all features have similar scales, which can be crucial for certain machine learning algorithms to work effectively. Here are some common techniques for feature normalization:

**1. Min-Max Scaling (Normalization):**
* Min-max scaling scales features to a specific range, typically [0, 1].

* It is calculated using the formula:

X_normalized = (X - X_min) / (X_max - X_min)

* This technique is sensitive to outliers and may not work well if your data contains extreme values.

**2. Z-Score Standardization (Standardization):**
* Z-score standardization scales features to have a mean of 0 and a standard deviation of 1.

* It is calculated using the formula:

X_standardized = (X - mean(X)) / std(X)

* It is less sensitive to outliers compared to min-max scaling and is commonly used in statistics and machine learning.

# Tuple Duplication:
A tuple is an ordered collection of elements, and each element can be of any data type. Tuples are similar to lists, but unlike lists, they are immutable, which means you cannot change their contents (add, remove, or modify elements) once they are created. Tuple duplication typically refers to creating a new tuple that contains the same elements as an existing tuple, essentially making a copy of the original tuple. You can duplicate a tuple in Python using various methods.
Handling duplicate tuples is essential for data cleaning and analysis. Pandas provides several methods to deal with duplicate tuples effectively:

**duplicated( ):** This method returns a boolean Series indicating whether each row is a duplicate of a previous row.

**drop_duplicates( ):** This method removes duplicate tuples from the DataFrame. By default, it keeps the first occurrence and removes subsequent duplicates.

**Paraneters**:

subset: Specifies columns to consider when detecting duplicates.

keep: Determines which duplicates to mark as True (first, last, or all occurrences).

inplace: Modifies the DataFrame in-place, if set to True.

ignore_index: Resets the index after removing duplicates, if set to True.

```python
df.drop_duplicates(subset=None)
```

# Remove duplicate rows in a pandas DataFrame based on a specific column, retaining the first occurrence of each duplicate and deleting subsequent occurrences.

```python
df.drop_duplicates(subset=['brand'],inplace=False,keep='first')
```

# Remove duplicate rows in a pandas DataFrame based on a specific column, keeping the last occurrence of each duplicate and deleting earlier occurrences.

```python
df.drop_duplicates(subset=['brand'],inplace=False,keep='last')
```

# Remove all duplicate rows in a pandas DataFrame without keeping any of the duplicate occurrences, using the keep=False parameter.

```python
df.drop_duplicates(subset=['brand'],inplace=False,keep=False)
```
