# plt.show()
In Python, plt.show() is a function from the matplotlib library that is used to display a plot or figure on the screen or in an interactive graphical user interface (GUI) window. It is typically used at the end of a script or code block to render and show the visual representation of the data created using matplotlib.



```python
# Display Basic Descriptive Data Summaries
print("Displaying Basic Descriptive Data Summaries:")
print(df.describe())

# Create a line plot showing the Total Profit for each month
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['YearMonth'] = df['ORDERDATE'].dt.to_period('M')
total_profit_by_month = df.groupby('YearMonth')['SALES'].sum()

# Convert Period objects to strings for plotting
total_profit_by_month.index = total_profit_by_month.index.strftime('%Y-%m')

plt.figure(figsize=(12, 6))
plt.plot(total_profit_by_month.index, total_profit_by_month.values, marker='o', color='purple')
plt.title('Total Profit Over The Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Create a figure with a specified size
plt.figure(figsize=(12, 6))

# Plot the Total Profit data with custom visual properties
plt.plot(
    total_profit_by_month.index,
    total_profit_by_month.values,
    color='blue',            # Set line color to blue
    marker='o',              # Use circular markers
    markerfacecolor='black', # Set marker face color to black
    linestyle='--',          # Use dashed line style
    linewidth=3              # Set line width to 3
)

# Set plot title and labels
plt.title('Total Profit Over The Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display grid lines on the plot
plt.grid(True)

# Show the plot
plt.show()
```

# Reading all the product sales data and show it  using a Multiline Plot

```python
# Convert the 'ORDERDATE' column to datetime objects
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Create a new 'YearMonth' column for grouping by month and year
df['YearMonth'] = df['ORDERDATE'].dt.to_period('M')

# Group the data by 'PRODUCTLINE' and 'YearMonth' to calculate total sales
product_sales = df.groupby(['PRODUCTLINE', 'YearMonth'])['SALES'].sum().unstack()

# Create a line plot for each product's sales over time
plt.figure(figsize=(12, 6))
for product in product_sales.columns:
    plt.plot(product_sales.index, product_sales[product], label=product)

plt.title('Product Sales Over The Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.show()
```

# Reading the Classic Cars sales data of each month and show it using a Scatter Plot

# plt.scatter()
In Python, plt.scatter() is a function provided by the matplotlib library that is specifically designed for crafting scatter plots. Scatter plots are a visualization technique employed to illustrate the correlation or association between two datasets by representing individual data points as dots on a two-dimensional plane.

```python
# Convert the 'ORDERDATE' column to datetime objects
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Create a new 'YearMonth' column for grouping by month
df['YearMonth'] = df['ORDERDATE'].dt.to_period('M')

# Extract sales data for 'Classic Cars'
classic_cars_sales = df[df['PRODUCTLINE'] == 'Classic Cars']

# Convert 'YearMonth' Periods to timestamps
classic_cars_sales['YearMonth'] = classic_cars_sales['YearMonth'].dt.to_timestamp()

# Create a scatter plot for Classic Cars sales over time
plt.figure(figsize=(10, 6))
plt.scatter(classic_cars_sales['YearMonth'], classic_cars_sales['SALES'], c='orange', marker='o')
plt.title('Classic Cars Sales Over Months (Scatter Plot)')
plt.xlabel('Month')
plt.ylabel('Classic Cars Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# Reading the Trucks and Buses and Classic Cars sales data and show it using the Bar Chart

```python
# Convert 'ORDERDATE' to a datetime object
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Create a 'YearMonth' column for grouping
df['YearMonth'] = df['ORDERDATE'].dt.to_period('M')

# Bar chart for Trucks and Buses and Classic Cars sales data
selected_products = ['Trucks and Buses', 'Classic Cars']
selected_data = df[df['PRODUCTLINE'].isin(selected_products)]

# Convert 'YearMonth' Periods to strings
selected_data['YearMonth'] = selected_data['YearMonth'].dt.strftime('%Y-%m')

# Pivot the data to prepare for the bar chart
pivot_data = selected_data.pivot_table(index='YearMonth', columns='PRODUCTLINE', values='SALES', aggfunc='sum')

# Define colors for the bars
bar_colors = ['yellow', 'red']

plt.figure(figsize=(10, 6))
pivot_data.plot(kind='bar', stacked=True, color=bar_colors)
plt.title('Sales of Trucks and Buses VS. Classic Cars Over Months (Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(selected_products)
plt.grid(True)
plt.show()
```

# Calculating the total sale data for last year for each product and showing it using a Pie Chart

# plt.pie()
In Python, the plt.pie() function, part of the matplotlib library, is employed for crafting pie charts. A pie chart is a circular diagram segmented into slices, with each slice denoting a fraction of the entire dataset. These charts serve as valuable tools for illustrating the breakdown of categorical data and displaying the relative proportions of each category in relation to the whole dataset.



```python
# Defining custom colors for the pie chart segments
custom_colors = ['skyblue', 'purple', 'lightskyblue', 'yellow', 'red']

# Creating a pie chart with custom colors
plt.figure(figsize=(8, 8))
last_year_sales = df[df['YearMonth'] >= '2004-05'].groupby('PRODUCTLINE')['SALES'].sum()

# Adding the custom colors to the pie chart
plt.pie(last_year_sales, labels=last_year_sales.index, autopct='%1.1f%%', startangle=140, colors=custom_colors)

plt.title('Last Year Total Sales Distribution by Product (Pie Chart)')
plt.axis('equal')
plt.show()
```

# Calculating the Sales Distribution by Product Category and showing it with the Box Plot

```python
# Create a figure with a specific size
plt.figure(figsize=(12, 6))

# Generate a box plot for sales data by product category
# The list comprehension extracts sales data for each unique product category
# The 'vert=False' parameter specifies horizontal box plots
plt.boxplot([df[df['PRODUCTLINE'] == product]['SALES'] for product in df['PRODUCTLINE'].unique()],
            labels=df['PRODUCTLINE'].unique(), vert=False)

# Set the title of the plot
plt.title('Sales Data Distribution by Product Category (Box Plot)')

# Label the x-axis
plt.xlabel('Sales')

# Enable grid lines on the plot
plt.grid(True)

# Display the plot
plt.show()
```
